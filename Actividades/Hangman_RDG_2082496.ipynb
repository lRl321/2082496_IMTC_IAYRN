Hangman Game: Create a Python program that allows a user to play the game of Hangman. The program should randomly select a word from a predefined list, and the player must guess letters one at a time until they guess the word correctly or run out of attempts.

```python
import random
from IPython.display import clear_output

# Lista de palabras para adivinar
palabras = ["rafael", "roberto", "rodrigo", "gissel", "dafne", "bruno", "jazmin", "orlando", "carlos"]

# Arreglo para el ahorcado
AHORCADO = ['''
      +---+
      |   |
          |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========''']

# Función para mostrar el estado actual del juego
def mostrar_juego(adivinanza, letras_adivinadas, intentos):
    clear_output(wait=True)  # Borra la salida anterior
    print("\nPalabra a adivinar: " + " ".join(adivinanza))
    print("Letras adivinadas: " + " ".join(letras_adivinadas))
    print("Intentos restantes: {}".format(intentos))
    print(AHORCADO[6 - intentos])

# Función para preguntar al jugador si quiere volver a jugar
def jugar_nuevamente():
    while True:
        respuesta = input("\n¿Quieres jugar de nuevo? (s/n): ").lower()
        if respuesta == 's':
            return True
        elif respuesta == 'n':
            return False
        else:
            print("Respuesta inválida. Por favor, ingresa 's' para sí o 'n' para no.")

# Función para jugar una partida completa
def jugar_partida():
    # Seleccionar una palabra al azar
    palabra_secreta = random.choice(palabras)

    # Inicializar variables
    intentos = 6
    letras_adivinadas = []
    adivinanza = ["_"] * len(palabra_secreta)

    print("¡Bienvenido al juego del ahorcado!")

    while True:
        mostrar_juego(adivinanza, letras_adivinadas, intentos)

        # Pedir al jugador que adivine una letra
        letra = input("\nAdivina una letra: ").lower()

        # Verificar si la letra ya fue adivinada
        if letra in letras_adivinadas:
            print("Ya adivinaste esa letra. Intenta con otra.")
            continue

        # Verificar si la letra está en la palabra secreta
        if letra in palabra_secreta:
            letras_adivinadas.append(letra)
            for i in range(len(palabra_secreta)):
                if palabra_secreta[i] == letra:
                    adivinanza[i] = letra
        else:
            letras_adivinadas.append(letra)
            intentos -= 1
            print("Letra incorrecta. Te quedan {} intentos.".format(intentos))

        # Verificar si el jugador ganó o perdió
        if "".join(adivinanza) == palabra_secreta:
            mostrar_juego(adivinanza, letras_adivinadas, intentos)
            print("\n¡Felicidades! Has adivinado la palabra correctamente: '{}'".format(palabra_secreta))
            break
        elif intentos == 0:
            mostrar_juego(adivinanza, letras_adivinadas, intentos)
            print("\n¡Fin del juego! La palabra era '{}'".format(palabra_secreta))
            break

# Jugar una partida y preguntar si quiere volver a jugar
while True:
    jugar_partida()
    if not jugar_nuevamente():
        break
```
